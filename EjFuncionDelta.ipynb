```python
import numpy as np #Libreria para dar sentido matemtico
from math import pi
import matplotlib.pyplot as plt #Libreria que permite ver gficos
from scipy import integrate #Permite resolver integrales

# Funcin cuadrtica.
def f1(x):
    return np.cos(x)
# Funcin lineal.
def f2(x):
    return np.cos(x)+4*np.sin(x)

#Condicin incial
x0=np.linspace(0*pi,2*pi,100)
x1=np.linspace(2*pi,5*pi,100)
x2=np.linspace(2*pi,5*pi,100)

# Configuracin del grfico
fig, axes=plt.subplots()
axes.plot(x0,f1(x0),'g')
axes.plot(x2,f1(x2),'--r')
axes.plot(x1,f2(x1),'--b')

plt.title('Ec. Dif. 2do Orden')
plt.xlabel('Tiempo(seg)')
plt.ylabel('Frecuencia')
plt.legend(['Sol. 1ra ED. "Posicin"','Sol. 2da ED. "Velocidad"'])
plt.grid( True )
plt.show()
```
