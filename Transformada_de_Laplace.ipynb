# **Transformadas de Laplace en Python**

 Python y Sympy, son herramientas muy poderosas, para resolver por ejemplo ecuaciones diferenciales.

Calculemos las transformadas de Laplace de algunas funciones típicas y comencemos con la más simple: la función constante. Usando Python para hacerlo:

```python
from sympy import *
import numpy as np

t=Symbol('t', real=True)
p=Symbol('p', real=True, positive=True)

#a=5
#b=6
#c=-8
A=-5
B=2
ω=5/2

f=A*cos(ω*t)+B*sin(ω*t)
#a+b*t+c*t**2
#3+t**5+sin(np.pi*t)
#a+b*t+c*t**2
#Acos(ωt)+Bsin(ωt)


F=Integral(f*exp(-p*t),(t,0,oo))
F
```

```python
F.doit()
```

Hay una manera más fácil en sympy de calcular la transformada de Laplace. Basta con utilizar la función dedicada:

```python
laplace_transform(f,t,p)
```

Tenga en cuenta que la salida de esa función no solo le proporciona la transformada de Laplace, sino también información sobre si el cálculo se realizó correctamente. La razón es que la integral es impropia, ya que nos integramos hasta el infinito. Y puede ser que la integral no converja. No es así en nuestro caso. Si solo quieres la solución, usa la opción: noconds=True

```python
laplace_transform(f,t,p,noconds=True)
```

```python
# importando modulos necesarios
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import sympy
from scipy import integrate

# imprimir con notación matemática.
sympy.init_printing(use_latex='mathjax')
```

```python
# Resolviendo ecuación diferencial
# defino las incognitas
x = sympy.Symbol('x')
y = sympy.Function('y')

# expreso la ecuacion
f = 6*x**2 - 3*x**2*(y(x))
sympy.Eq(y(x).diff(x), f)
```

```python
# Resolviendo la ecuación
sympy.dsolve(y(x).diff(x) - f)
```
